**Идентификаторы и Метаданные:**

1.  `#Pdb` (`object`): Идентификатор структуры белкового комплекса в Protein Data Bank (PDB). Указывает, для какой именно структуры проводился анализ или эксперимент. (348 уникальных значений).
2.  `Reference` (`object`): Библиографическая ссылка (часто PubMed ID или DOI) на статью, откуда были взяты экспериментальные данные.
3.  `Protein 1` (`object`): Название или идентификатор первого белка в комплексе.
4.  `Protein 2` (`object`): Название или идентификатор второго белка в комплексе.
5.  `Hold_out_type` (`object`): Тип данных в отложенной выборке (hold-out set), использованной для валидации в оригинальной статье SKEMPI. Много пропусков (~53%), вероятно, не критичен для базовой модели.
6.  `Hold_out_proteins` (`object`): Белки, включенные в отложенную выборку.
7.  `Notes` (`object`): Дополнительные текстовые примечания к записи данных. Много пропусков.
8.  `SKEMPI version` (`int64`): Версия базы данных SKEMPI, из которой взята запись.

**Описание Мутации:**

9.  `Mutation(s)_PDB` (`object`): Описание мутации (или мутаций) в формате, близком к используемому в PDB файлах (например, `RA117C` означает Аргинин (R) в позиции 117 цепи A заменен на Цистеин (C)).
10. `Mutation(s)_cleaned` (`object`): Очищенная или стандартизированная версия описания мутации. Часто используется как основа для парсинга.
11. `iMutation_Location(s)` (`object`): Категория местоположения мутации относительно интерфейса связывания белков. Ключевой признак. Основные категории:
    *   `COR`: Core (ядро интерфейса)
    *   `RIM`: Rim (край интерфейса)
    *   `SUP`: Support (поддерживающие остатки, не в прямом контакте)
    *   `SUR`: Surface (поверхность белка, вне интерфейса)
    *   `INT`: Interior (внутри белка, вне интерфейса)
    Могут быть комбинации для множественных мутаций (487 уникальных значений).

**Экспериментальные Условия:**

12. `Temperature` (`float64`): Температура (в Кельвинах, K), при которой проводился эксперимент по измерению связывания.
13. `Method` (`object`): Экспериментальный метод, использованный для измерения параметров связывания (например, SPR - поверхностный плазмонный резонанс, ITC - изотермическая титрационная калориметрия, FL - флуоресценция). (20 уникальных значений).

**Параметры Связывания (Аффинность и Кинетика):**

*   `_wt` означает "wild-type" (дикий тип, исходный белок).
*   `_mut` означает "mutant" (мутантный белок).
*   `_parsed` обычно означает числовое значение, извлеченное из строкового представления (которое могло содержать знаки `>`, `<`, `~`). Эти `_parsed` колонки обычно используются для расчетов.

14. `Affinity_mut (M)` (`object`): Исходная запись аффинности связывания для *мутантного* комплекса (часто Kd - константа диссоциации), возможно, в виде строки с указанием предела (> / <). Единицы: Моль (M).
15. `Affinity_mut_parsed` (`float64`): Числовое значение аффинности (Kd) *мутанта* в Молях (M). Низкое значение Kd означает сильное связывание. (~4% пропусков).
16. `Affinity_wt (M)` (`object`): Исходная запись аффинности связывания для *дикого типа*. Единицы: Моль (M).
17. `Affinity_wt_parsed` (`float64`): Числовое значение аффинности (Kd) *дикого типа* в Молях (M). (Очень мало пропусков).
18. `Kd_mut` (`float64`): Константа диссоциации для *мутанта* (вероятно, то же самое, что `Affinity_mut_parsed`). Единицы: Моль (M).
19. `Kd_wt` (`float64`): Константа диссоциации для *дикого типа* (вероятно, то же самое, что `Affinity_wt_parsed`). Единицы: Моль (M).
20. `kon_mut (M^(-1)s^(-1))` (`float64`): Константа скорости ассоциации (связывания) для *мутанта*. Единицы: M⁻¹s⁻¹. (Много пропусков, ~74%).
21. `kon_mut_parsed` (`float64`): Числовое значение `kon_mut`.
22. `kon_wt (M^(-1)s^(-1))` (`float64`): Константа скорости ассоциации для *дикого типа*. Единицы: M⁻¹s⁻¹. (Много пропусков, ~74%).
23. `kon_wt_parsed` (`float64`): Числовое значение `kon_wt`.
24. `koff_mut (s^(-1))` (`object`): Исходная запись константы скорости диссоциации (распада комплекса) для *мутанта*. Единицы: s⁻¹. (Много пропусков, ~74%).
25. `koff_mut_parsed` (`float64`): Числовое значение `koff_mut`.
26. `koff_wt (s^(-1))` (`object`): Исходная запись константы скорости диссоциации для *дикого типа*. Единицы: s⁻¹. (Много пропусков, ~74%).
27. `koff_wt_parsed` (`float64`): Числовое значение `koff_wt`.

**Термодинамические Параметры (Редко Заполнены):**

28. `dH_mut (kcal mol^(-1))` (`float64`): Изменение энтальпии при связывании для *мутанта*. Единицы: ккал/моль. (Очень много пропусков, ~94%).
29. `dH_wt (kcal mol^(-1))` (`float64`): Изменение энтальпии при связывании для *дикого типа*. Единицы: ккал/моль. (Очень много пропусков, ~94%).
30. `dS_mut (cal mol^(-1) K^(-1))` (`float64`): Изменение энтропии при связывании для *мутанта*. Единицы: кал/(моль·К). (Очень много пропусков, ~94%).
31. `dS_wt (cal mol^(-1) K^(-1))` (`float64`): Изменение энтропии при связывании для *дикого типа*. Единицы: кал/(моль·К). (Очень много пропусков, ~94%).

**Целевые Переменные:**

32. `ddG` (`float64`): **Изменение свободной энергии Гиббса при связывании** в результате мутации (ΔG_мутант - ΔG_дикий_тип). Рассчитывается как `RT * ln(Kd_mut / Kd_wt)`, где R - газовая постоянная, T - температура в Кельвинах. Единицы: ккал/моль.
    *   `ddG < 0`: Мутация стабилизирует комплекс (связывание сильнее).
    *   `ddG > 0`: Мутация дестабилизирует комплекс (связывание слабее).
    *   `ddG ≈ 0`: Мутация имеет нейтральный эффект.
    (Имеет ~4% пропусков).
33. `ddG_sign` (`float64`): Знак `ddG`, использовался в EDA для анализа баланса классов (-1: отриц., 0: нейтр., +1: полож.). В коде моделирования мы обычно создаем **бинарную версию** этой переменной (например, 0 для ddG < 0, 1 для ddG >= 0), которая и является **целевой переменной для задачи классификации**.

Этот детальный разбор помогает понять природу данных и выбрать наиболее релевантные и полные столбцы для построения модели.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Настройка стиля визуализаций
plt.style.use('seaborn')
sns.set_palette('husl')

# Загрузка данных
df = pd.read_csv('skempi_v2.csv', sep=';')

# 1. Обработка температуры
df['Temperature'] = df['Temperature'].str.replace(r'\(assumed\)', '', regex=True)
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')
median_temp = df['Temperature'].median()
df['Temperature'].fillna(median_temp, inplace=True)

# 2. Вычисление ddG
R = 1.987  # кал/(моль·К)
df['Kd_mut'] = df['Affinity_mut_parsed']
df['Kd_wt'] = df['Affinity_wt_parsed']
df['ddG'] = np.where(
    (df['Kd_mut'].notnull()) & (df['Kd_wt'].notnull()) & (df['Kd_wt'] != 0),
    -R * df['Temperature'] * np.log(df['Kd_mut'] / df['Kd_wt']) / 1000,  # в ккал/моль
    np.nan
)
df['ddG_sign'] = np.sign(df['ddG']).map({-1: 'Negative', 0: 'Neutral', 1: 'Positive'})  # Для читаемости

# 3. Извлечение позиции мутации
df['Position'] = df['Mutation(s)_cleaned'].str[1:-1].str.extract(r'(\d+)').astype(float)

# 4. Обзор данных
print("Общая информация о данных:")
print(df.info())
print("\nПропуски в данных:")
print(df.isnull().sum())

# 5. Анализ целевой переменной (ddG)
plt.figure(figsize=(10, 6))
sns.histplot(df['ddG'].dropna(), kde=True, bins=50)
plt.title('Распределение ddG (ккал/моль)')
plt.xlabel('ddG')
plt.ylabel('Частота')
plt.savefig('ddg_distribution.png')
plt.show()

# Баланс классов для ddG_sign
print("\nБаланс классов ddG_sign:")
class_counts = df['ddG_sign'].value_counts(normalize=True) * 100
print(class_counts)
plt.figure(figsize=(6, 4))
sns.countplot(x='ddG_sign', data=df, order=['Negative', 'Neutral', 'Positive'])
plt.title('Распределение знака ddG')
plt.xlabel('Знак ddG')
plt.ylabel('Количество')
plt.savefig('ddg_sign_distribution.png')
plt.show()

# 6. Анализ позиции мутации
plt.figure(figsize=(10, 6))
sns.histplot(df['Position'].dropna(), kde=True, bins=50)
plt.title('Распределение позиций мутаций')
plt.xlabel('Позиция')
plt.ylabel('Частота')
plt.savefig('position_distribution.png')
plt.show()

# Связь позиции с ddG
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Position', y='ddG', data=df)
plt.title('Зависимость ddG от позиции мутации')
plt.xlabel('Позиция')
plt.ylabel('ddG (ккал/моль)')
plt.savefig('position_ddg_scatter.png')
plt.show()

# 7. Детальный анализ iMutation_Location(s)
print("\nУникальные значения iMutation_Location(s) и их частоты:")
location_counts = df['iMutation_Location(s)'].value_counts()
print(location_counts)

# Разделение на основные и составные категории
basic_locations = ['COR', 'INT', 'SUP', 'RIM', 'SUR']
df['Is_Basic_Location'] = df['iMutation_Location(s)'].apply(lambda x: x in basic_locations)
basic_df = df[df['Is_Basic_Location']]

# Boxplot для основных локаций
plt.figure(figsize=(8, 6))
sns.boxplot(x='iMutation_Location(s)', y='ddG', data=basic_df, order=basic_locations)
plt.title('ddG для основных местоположений мутаций')
plt.xlabel('Местоположение мутации')
plt.ylabel('ddG (ккал/моль)')
plt.xticks(rotation=45)
plt.savefig('basic_location_ddg_boxplot.png')
plt.show()

# 8. Анализ категориальных признаков
plt.figure(figsize=(8, 5))
sns.countplot(y='iMutation_Location(s)', data=df, order=df['iMutation_Location(s)'].value_counts().index[:10])
plt.title('Топ-10 местоположений мутаций')
plt.xlabel('Количество')
plt.ylabel('Местоположение мутации')
plt.savefig('top10_locations.png')
plt.show()

plt.figure(figsize=(8, 5))
sns.countplot(y='Method', data=df, order=df['Method'].value_counts().index)
plt.title('Распределение методов экспериментов')
plt.xlabel('Количество')
plt.ylabel('Метод')
plt.savefig('method_distribution.png')
plt.show()

# 9. Анализ числовых признаков
df['log_Kd_mut'] = np.log10(df['Kd_mut'].replace(0, np.nan))
df['log_Kd_wt'] = np.log10(df['Kd_wt'].replace(0, np.nan))

plt.figure(figsize=(10, 6))
sns.histplot(df['log_Kd_mut'].dropna(), kde=True, label='log(Kd_mut)', alpha=0.5, bins=50)
sns.histplot(df['log_Kd_wt'].dropna(), kde=True, label='log(Kd_wt)', alpha=0.5, bins=50)
plt.title('Распределение логарифмов Kd (мутант и WT)')
plt.xlabel('log10(Kd)')
plt.ylabel('Частота')
plt.legend()
plt.savefig('log_kd_distribution.png')
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df['Temperature'], kde=True, bins=20)
plt.title('Распределение температуры')
plt.xlabel('Температура (K)')
plt.ylabel('Частота')
plt.savefig('temperature_distribution.png')
plt.show()

# 10. Корреляционный анализ
numeric_cols = ['Affinity_mut_parsed', 'Affinity_wt_parsed', 'kon_mut_parsed', 'kon_wt_parsed', 
                'koff_mut_parsed', 'koff_wt_parsed', 'Temperature', 'ddG', 'Position']
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Корреляционная матрица числовых признаков')
plt.savefig('correlation_matrix.png')
plt.show()

# 11. Анализ пропусков
print("\nПроцент пропусков в числовых столбцах:")
for col in numeric_cols + ['dH_mut (kcal mol^(-1))', 'dH_wt (kcal mol^(-1))', 
                           'dS_mut (cal mol^(-1) K^(-1))', 'dS_wt (cal mol^(-1) K^(-1))']:
    missing = df[col].isnull().mean() * 100
    print(f"{col}: {missing:.2f}%")

# 12. Анализ мутаций
df['Mutation_from'] = df['Mutation(s)_cleaned'].str[0]
df['Mutation_to'] = df['Mutation(s)_cleaned'].str[-1]

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.countplot(y='Mutation_from', data=df, order=df['Mutation_from'].value_counts().index)
plt.title('Исходные аминокислоты')
plt.xlabel('Количество')
plt.ylabel('Аминокислота')

plt.subplot(1, 2, 2)
sns.countplot(y='Mutation_to', data=df, order=df['Mutation_to'].value_counts().index)
plt.title('Мутированные аминокислоты')
plt.xlabel('Количество')
plt.ylabel('Аминокислота')
plt.tight_layout()
plt.savefig('mutation_amino_acids.png')
plt.show()

# 13. Выводы
print("\nОсновные выводы EDA:")
print("- Датасет: 7085 записей, 29 столбцов. Пропуски: dH/dS (94%), kon/koff (74%), Hold_out_type (53%).")
print("- ddG: нормальное распределение, пропуски (~4%) из-за Affinity. Баланс классов ddG_sign сбалансирован.")
print("- Позиции мутаций: распределение неравномерное, слабая связь с ddG.")
print("- iMutation_Location(s): основные категории (COR, INT, SUP) влияют на ddG, составные требуют обработки.")
print("- Числовые признаки (log_Kd_mut, log_Kd_wt, Temperature) информативны, требуют нормализации.")
print("- Мутации: R, L чаще мутируют, полезно для feature engineering (физико-химические свойства).")
print("- Рекомендации: удалить строки с пропусками в Affinity, исключить dH/dS, закодировать категориальные признаки.")
```

### EDA Анализ (Исследовательский Анализ Данных)

На основе предоставленных данных и графиков EDA для датасета SKEMPI 2.0 можно сделать следующие ключевые выводы:

Основная цель – предсказать знак изменения свободной энергии связывания (`ddG`), которое показывает, стабилизирует (`ddG < 0`) или дестабилизирует (`ddG > 0`) мутация взаимодействие.
2.  **Целевая Переменная (`ddG`/`ddG_sign`):**
    *   Распределение `ddG` имеет пик около нуля, что ожидаемо, но также длинные хвосты, указывающие на значимые эффекты некоторых мутаций.
    *   При создании бинарной целевой переменной (`ddG_sign`: 0 для ddG < 0, 1 для ddG >= 0) наблюдается **сильный дисбаланс классов** (примерно 76% класс 0, 24% класс 1). Это критически важно для выбора моделей и метрик оценки.
    *   Около 4% записей имеют пропуски в `ddG` и должны быть удалены.
3.  **Ключевые Предикторы:**
    *   **Информация о Мутации:** `WT_AA` (исходная АК), `MUT_AA` (мутантная АК), `Position` (позиция мутации) являются важными категориальными и числовыми признаками.

    *   **Условия Эксперимента:** `Temperature` (числовой, мало вариативен) и `Method` (категориальный) могут внести вклад.
4.  **Исключенные Признаки:**
    *   Кинетические (`kon`, `koff`) и термодинамические (`dH`, `dS`) параметры исключены из-за >70% пропусков.
    *   Аффинности (`Affinity_*`, `Kd_*`) **нельзя использовать** как признаки для предсказания `ddG`, так как `ddG` вычисляется из них же (приведет к **утечке данных**).
5.  **Выводы EDA:** Датасет позволяет строить модель, но требует внимания к дисбалансу классов и сложной обработке признака локации. Слабые линейные корреляции указывают на важность категориальных признаков и нелинейных зависимостей.

### Возможные Пути Использования в Фарм Промышленности (Фокус на Моноклональных Антителах)

Модель, предсказывающая знак (или величину) `ddG` для белок-белковых взаимодействий, может быть ценным инструментом на **ранних стадиях разработки моноклональных антител (мАТ)**, в первую очередь как **средство для быстрой фильтрации и приоритизации**:

1.  **Созревание Аффинности:**
    *   **Задача:** Усилить связывание мАТ с антигеном (`ddG < 0`).
    *   **Использование:** *In silico* скрининг библиотек мутаций (особенно в CDR) для отбора кандидатов с предсказанным `ddG < 0` и отсеивания тех, для которых `ddG > 0`. Это экономит ресурсы на экспериментальную проверку.
2.  **Гуманизация / Деиммунизация:**
    *   **Задача:** Снизить иммуногенность мАТ, сохраняя аффинность.
    *   **Использование:** Оценка предлагаемых мутаций (в FR) для выявления тех, что могут **ослабить связывание (`ddG > 0`)**, помогая избежать потери функции.
3.  **Оценка Кандидатов и Генерация Гипотез:**
    *   **Задача:** Понять, какие регионы (CDR vs FR, ядро vs периферия интерфейса) и типы замен наиболее критичны для связывания.
    *   **Использование:** Анализ предсказаний модели и важности признаков может подсказать, какие мутации с наибольшей вероятностью повлияют на аффинность, направляя дальнейшие эксперименты.
4.  **Предсказание Устойчивости (Опосредованно):**
    *   **Задача:** Оценить, как мутации в антигене могут повлиять на связывание с мАТ.
    *   **Использование:** Предсказание знака ddG для мутаций в антигене может дать ранний сигнал о возможном "ускользании" от антитела (хотя модель обучена на разных PPI).

**Ограничения:** Текущая простая модель (предсказание только знака, простые признаки) будет иметь ограниченную точность. Для принятия критических решений необходима более точная модель, предсказывающая **величину** ddG и использующая более **релевантные признаки**.

### Подробный Дальнейший План по Реализации

    *   Загрузить данные `skempi_v2.0.csv` с помощью `pandas`, указав правильный разделитель (`;`) и символ комментария (`#`).
    *   Удалить строки, где значение в столбце `ddG` отсутствует (`dropna(subset=['ddG'])`).
    *   **Целевая Переменная:**
        *   **Вариант А (Классификация):** Создать бинарный столбец `ddG_sign = (df['ddG'] >= 0).astype(int)`.
        *   **Вариант Б (Регрессия - Рекомендуется):** Использовать столбец `ddG` напрямую как целевую переменную `y`.
2.  **Инженерия Признаков:**
    *   **Парсинг Мутации:** Извлечь `WT_AA`, `MUT_AA`, `Position` из `Mutation(s)_cleaned` или `Mutation(s)_PDB` с помощью регулярных выражений. Обработать ошибки парсинга.
    *   **Обработка Локации (`iMutation_Location(s)`):** Создать бинарные флаги для основных категорий (`COR`, `RIM`, `SUR`, `SUP`, `INT`). Например: `df['loc_COR'] = df['iMutation_Location(s)'].str.contains('COR').astype(int)`.
    *   **Физико-Химические Признаки (ΔСвойства):**
        *   Создать словарь со свойствами аминокислот (объем, гидрофобность Kyte-Doolittle, заряд и т.д.).
        *   Для каждой строки рассчитать разницу свойств между MUT и WT аминокислотами (например, `delta_volume`, `delta_hydrophobicity_kd`). Добавить их как новые столбцы.
3.  **Выбор Признаков (`X`):**
    *   Категориальные: `WT_AA`, `MUT_AA`, `Method`.
    *   Числовые: `Position`, `Temperature`, `delta_volume`, `delta_hydrophobicity_kd` (и другие Δ-признаки).
    *   Бинарные: `loc_COR`, `loc_RIM`, `loc_SUR`, `loc_SUP`, `loc_INT`.
4.  **Препроцессинг (внутри `Pipeline`):**
    *   Создать `ColumnTransformer`.
    *   Применить `OneHotEncoder(handle_unknown='ignore')` к категориальным признакам.
    *   Применить `StandardScaler()` к числовым признакам (включая Δ-признаки).
    *   Бинарные флаги локации можно передать без изменений (`'passthrough'`) или также масштабировать.
5.  **Выбор и Обучение Модели:**
    *   Разделить данные на обучающую и тестовую выборки (`train_test_split`), используя `stratify=y` (особенно важно для классификации из-за дисбаланса).
    *   **Модели:**
        *   **Классификация:** Начать с `RandomForestClassifier` или `LGBMClassifier`. **Обязательно** использовать параметр `class_weight='balanced'` для борьбы с дисбалансом.
        *   **Регрессия:** Использовать `RandomForestRegressor` или `LGBMRegressor`.
    *   Обучить пайплайн (препроцессинг + модель) на обучающей выборке.
6.  **Оценка Модели:**
    *   **Кросс-Валидация:** Использовать `cross_val_score` или `cross_validate` с `StratifiedKFold` (для классификации) или `KFold` (для регрессии) на *всем* пайплайне для получения надежной оценки.
    *   **Метрики:**
        *   Классификация: AUC ROC, F1-score (особенно для минорного класса), Precision, Recall, Confusion Matrix. **Избегать Accuracy.**
        *   Регрессия: RMSE, MAE, R² (коэффициент детерминации), Коэффициент корреляции Пирсона/Спирмена.
    *   **Сравнение с Baseline:** Сравнить результаты с `DummyClassifier` / `DummyRegressor`.
7.  **Анализ и Интерпретация:**
    *   Проанализировать важность признаков с помощью `.feature_importances_`, `permutation_importance` или библиотеки `shap`. Определить, какие факторы (тип АК, локация, физхим свойства) наиболее влияют на предсказание.
8.  **Итерации:** На основе результатов оценки и анализа важности признаков, при необходимости вернуться к шагам 2-5 для улучшения признаков или настройки/смены модели.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
import os

warnings.filterwarnings('ignore')

# --- Настройки стиля - Единая тема, Контуры, Высокие боксплоты ---
plt.style.use('seaborn-v0_8-white') # Чистый фон

# Глобальные параметры для шрифтов и осей
plt.rcParams.update({
    'font.family': 'sans-serif',
    'font.size': 12,
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    'legend.fontsize': 11,
    'figure.dpi': 300,
    'savefig.dpi': 300,
    'savefig.format': 'png',
    'savefig.bbox': 'tight',
    'axes.linewidth': 1.0,
    'axes.edgecolor': 'black',
    'xtick.major.size': 5,
    'ytick.major.size': 5,
    'xtick.major.width': 1.0,
    'ytick.major.width': 1.0,
})

# --- Определение Единой Цветовой Темы (на основе палитры "mako") ---
base_palette = sns.color_palette("mako", n_colors=6) # Берем 6 цветов из палитры

primary_color = base_palette[3]    # Основной цвет (средний тон)
secondary_color = base_palette[1]  # Вспомогательный (светлее)
tertiary_color = base_palette[5]   # Еще один оттенок (темнее)
accent_color = 'firebrick'         # Яркий акцент для линий нуля
heatmap_cmap = 'mako'              # Используем ту же палитру для heatmap
boxplot_palette = sns.color_palette("mako", n_colors=5) # Палитра для 5 категорий локаций
aa_color1 = base_palette[2]        # Цвет для WT AA
aa_color2 = base_palette[4]        # Цвет для Mut AA
stripplot_fill_color = base_palette[1] # Цвет заливки точек на стрипплоте (светлый)

# Контур для всех элементов
contour_color = 'black'
bar_contour_lw = 1.0
point_contour_lw = 0.5 # Тоньше контур для точек

# Папка для сохранения графиков
output_dir = 'eda_plots_unified_theme_png' # Новое имя папки
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# --- Загрузка и предобработка данных (без изменений) ---
try:
    df = pd.read_csv('skempi_v2.csv', sep=';')
except FileNotFoundError:
    print("Ошибка: Файл skempi_v2.csv не найден. Пожалуйста, поместите его в ту же директорию.")
    exit()

# 1. Обработка температуры
df['Temperature'] = df['Temperature'].str.replace(r'\(assumed\)', '', regex=True)
df['Temperature'] = pd.to_numeric(df['Temperature'], errors='coerce')
median_temp = df['Temperature'].median()
df['Temperature'].fillna(median_temp, inplace=True)

# 2. Вычисление ddG
R = 1.987e-3
df['Kd_mut'] = df['Affinity_mut_parsed']
df['Kd_wt'] = df['Affinity_wt_parsed']
mask = (df['Kd_mut'].notnull()) & (df['Kd_wt'].notnull()) & (df['Kd_wt'] > 0) & (df['Kd_mut'] > 0) & (df['Temperature'] > 0)
df['ddG'] = np.nan
df.loc[mask, 'ddG'] = -R * df.loc[mask, 'Temperature'] * np.log(df.loc[mask, 'Kd_mut'] / df.loc[mask, 'Kd_wt'])

df_clean = df.dropna(subset=['ddG']).copy()
df_clean = df_clean[df_clean['ddG'] != 0]
df_clean['ddG_sign_class'] = (df_clean['ddG'] < 0).astype(int)
df_clean['ddG_sign_label'] = df_clean['ddG'].apply(lambda x: 'Negative (< 0)' if x < 0 else 'Positive (> 0)')


# 3. Извлечение позиции мутации
df_clean['Position'] = df_clean['Mutation(s)_cleaned'].str[1:-1].str.extract(r'(\d+)').astype(float)

# 4. Обзор данных
print("Размер датасета после очистки ddG:", df_clean.shape)


# --- Визуализации в единой цветовой теме ---

# 5. Анализ целевой переменной (ddG)
plt.figure(figsize=(6, 4))
sns.histplot(df_clean['ddG'], kde=True, bins=50, color=primary_color, alpha=0.7,
             edgecolor=contour_color, linewidth=bar_contour_lw)
plt.title('Distribution of Binding Energy Change (ΔΔG)')
plt.xlabel('ΔΔG (kcal/mol)')
plt.ylabel('Frequency')
plt.axvline(0, color=accent_color, linestyle='--', linewidth=1.5, label='ΔΔG = 0')
plt.legend(frameon=False)
plt.savefig(os.path.join(output_dir, 'ddg_distribution.png'))
plt.close()

# Баланс классов для ddG_sign_label
print("\nClass balance for ΔΔG sign:")
class_counts = df_clean['ddG_sign_label'].value_counts(normalize=True) * 100
print(class_counts)
plt.figure(figsize=(4, 4))
# Используем основной и вторичный цвета для двух категорий
count_colors = [secondary_color, primary_color] # Негативный - светлее, Позитивный - темнее
sns.countplot(x='ddG_sign_label', data=df_clean, order=['Negative (< 0)', 'Positive (> 0)'],
              palette=count_colors, edgecolor=contour_color, linewidth=bar_contour_lw)
plt.title('ΔΔG Sign Distribution')
plt.xlabel('Sign of ΔΔG')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.savefig(os.path.join(output_dir, 'ddg_sign_distribution.png'))
plt.close()

# 6. Анализ позиции мутации
plt.figure(figsize=(6, 4))
sns.histplot(df_clean['Position'].dropna(), kde=False, bins=50, color=primary_color, alpha=0.7,
             edgecolor=contour_color, linewidth=bar_contour_lw)
plt.title('Distribution of Mutation Positions')
plt.xlabel('Residue Position')
plt.ylabel('Frequency')
plt.savefig(os.path.join(output_dir, 'position_distribution.png'))
plt.close()

# Связь позиции с ddG (Scatter plot)
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Position', y='ddG', data=df_clean, alpha=0.5, s=25,
                color=primary_color, edgecolor=contour_color, linewidth=point_contour_lw) # Контур маркера
plt.title('ΔΔG vs. Mutation Position')
plt.xlabel('Residue Position')
plt.ylabel('ΔΔG (kcal/mol)')
plt.axhline(0, color=accent_color, linestyle='--', linewidth=1.5)
plt.savefig(os.path.join(output_dir, 'position_ddg_scatter.png'))
plt.close()

# 7. Детальный анализ iMutation_Location(s) - ВЫСОКИЙ БОКСПЛОТ
print("\nUnique iMutation_Location(s) values and counts:")
location_counts = df_clean['iMutation_Location(s)'].value_counts()
print(location_counts.head(10))

basic_locations = ['COR', 'SUR', 'RIM', 'INT', 'SUP']
basic_df = df_clean[df_clean['iMutation_Location(s)'].isin(basic_locations)].copy()
basic_df['iMutation_Location(s)'] = pd.Categorical(basic_df['iMutation_Location(s)'], categories=basic_locations, ordered=True)

plt.figure(figsize=(7, 7)) # УВЕЛИЧИВАЕМ ВЫСОТУ (было 5)
# Boxplot с основной палитрой темы
sns.boxplot(x='iMutation_Location(s)', y='ddG', data=basic_df,
            palette=boxplot_palette, showfliers=False,
            linewidth=bar_contour_lw + 0.5) # Чуть толще линии боксов
# Stripplot с контурами точек
sns.stripplot(x='iMutation_Location(s)', y='ddG', data=basic_df,
              color=stripplot_fill_color, # Светлая заливка
              edgecolor=contour_color,    # Черный контур
              linewidth=point_contour_lw, # Толщина контура точки
              alpha=0.6, size=4, jitter=True) # Немного увеличим точки и прозрачность
plt.title('ΔΔG by Core Mutation Location')
plt.xlabel('Mutation Location')
plt.ylabel('ΔΔG (kcal/mol)')
plt.xticks(rotation=0)
plt.axhline(0, color=accent_color, linestyle='--', linewidth=1.5)
plt.savefig(os.path.join(output_dir, 'basic_location_ddg_boxplot_stripplot.png'))
plt.close()

# 8. Анализ категориальных признаков (Топ-10 локаций)
top_locations = location_counts.index[:10]
plt.figure(figsize=(6, 5))
sns.countplot(y='iMutation_Location(s)', data=df_clean[df_clean['iMutation_Location(s)'].isin(top_locations)],
              order=top_locations, color=primary_color, # Основной цвет
              edgecolor=contour_color, linewidth=bar_contour_lw)
plt.title('Top 10 Mutation Locations')
plt.xlabel('Count')
plt.ylabel('Location')
plt.savefig(os.path.join(output_dir, 'top10_locations.png'))
plt.close()

# 9. Анализ числовых признаков (логарифмы Kd)
df_clean['log_Kd_mut'] = np.log10(df_clean['Kd_mut'].replace(0, np.nan))
df_clean['log_Kd_wt'] = np.log10(df_clean['Kd_wt'].replace(0, np.nan))

plt.figure(figsize=(6, 4))
# Используем разные оттенки из основной палитры
sns.histplot(df_clean['log_Kd_mut'].dropna(), kde=True, label='Mutant', alpha=0.7, bins=50,
             color=tertiary_color, edgecolor=contour_color, linewidth=bar_contour_lw) # Темный оттенок
sns.histplot(df_clean['log_Kd_wt'].dropna(), kde=True, label='Wild Type', alpha=0.7, bins=50,
             color=secondary_color, edgecolor=contour_color, linewidth=bar_contour_lw) # Светлый оттенок
plt.title('Distribution of log$_{10}$(K$_d$)')
plt.xlabel('log$_{10}$(K$_d$) (M)')
plt.ylabel('Frequency')
plt.legend(title='Protein', frameon=False)
plt.savefig(os.path.join(output_dir, 'log_kd_distribution.png'))
plt.close()

# 10. Корреляционный анализ
numeric_cols = ['Affinity_mut_parsed', 'Affinity_wt_parsed', 'kon_mut_parsed', 'kon_wt_parsed',
                'koff_mut_parsed', 'koff_wt_parsed', 'Temperature', 'ddG', 'Position']
corr_df = df_clean[numeric_cols].dropna()
corr_matrix = corr_df.corr()

plt.figure(figsize=(8, 7))
sns.heatmap(corr_matrix, annot=True, cmap=heatmap_cmap, vmin=-1, vmax=1, fmt=".2f", # Используем основную палитру
            linewidths=.5, linecolor='lightgray', cbar_kws={"shrink": .8})
plt.title('Correlation Matrix of Numeric Features')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.savefig(os.path.join(output_dir, 'correlation_matrix.png'))
plt.close()

# 12. Анализ мутаций (исходные и мутированные AA)
df_clean['Mutation_from'] = df_clean['Mutation(s)_cleaned'].str[0]
df_clean['Mutation_to'] = df_clean['Mutation(s)_cleaned'].str[-1]
aa_order = sorted(list(set(df_clean['Mutation_from']) | set(df_clean['Mutation_to'])))

fig, axes = plt.subplots(1, 2, figsize=(9, 6), sharey=True)

sns.countplot(y='Mutation_from', data=df_clean, order=aa_order, ax=axes[0],
              color=aa_color1, edgecolor=contour_color, linewidth=bar_contour_lw) # Цвет 1 для WT
axes[0].set_title('Original Amino Acids (WT)')
axes[0].set_xlabel('Count')
axes[0].set_ylabel('Amino Acid')

sns.countplot(y='Mutation_to', data=df_clean, order=aa_order, ax=axes[1],
              color=aa_color2, edgecolor=contour_color, linewidth=bar_contour_lw) # Цвет 2 для Mut
axes[1].set_title('Mutated Amino Acids')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('')

fig.suptitle('Amino Acid Mutation Frequencies', y=1.02)
plt.tight_layout(rect=[0, 0, 1, 1])
plt.savefig(os.path.join(output_dir, 'mutation_amino_acids.png'))
plt.close()

print(f"\nГрафики в единой теме сохранены в формате PNG в папку: {output_dir}")
```
